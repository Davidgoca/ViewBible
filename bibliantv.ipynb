```python
import pandas as pd
import spacy
from io import StringIO
import requests

# URL del archivo CSV en GitHub
url = "https://raw.githubusercontent.com/Davidgoca/biblia/main/biblia_ntv_.csv"

# Descargar el contenido del archivo CSV
response = requests.get(url)
csv_data = StringIO(response.text)

# Leer el archivo CSV en un DataFrame de pandas
df = pd.read_csv(csv_data)

# Cargar el modelo de spaCy
nlp = spacy.load("es_core_news_sm")  # Ajusta según tu modelo spaCy y tu idioma

# Procesar cada texto en la columna "texto"
for index, row in df.iterrows():
    doc = nlp(row['texto'])

    # Realizar operaciones de procesamiento de texto con spaCy
    # Por ejemplo, imprimir entidades nombradas
    print(f"Texto {index + 1}:")
    for entidad in doc.ents:
        print(f"{entidad.text}: {entidad.label_}")

    print("\n")
```
