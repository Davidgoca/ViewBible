```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Agregar nodos
G.add_node("Génesis", testamento="Antiguo Testamento")
G.add_node("Capitulo 1")
G.add_node("Versiculo 1", texto="En el principio creó Dios los cielos y la tierra.")
G.add_node("Moisés", rol="Profeta")
G.add_node("Éxodo", fecha="Siglo XIII a.C.")

# Agregar relaciones
G.add_edge("Génesis", "Capitulo 1", relationship="TIENE")
G.add_edge("Capitulo 1", "Versiculo 1", relationship="CONTIENE")
G.add_edge("Moisés", "Éxodo", relationship="PARTICIPA_EN")

# Visualizar el grafo
pos = nx.spring_layout(G)
labels = {node: node for node in G.nodes}

plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, labels=labels, font_size=8, node_color="lightblue", node_size=2000, font_color="black", font_weight="bold", arrowsize=20)
plt.title("Modelo de Base de Datos de Grafo para la Biblia")
plt.show()
```
